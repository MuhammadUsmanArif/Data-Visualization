```python
import matplotlib.pyplot as plt
squares = [1,4,9,16,25]

plt.plot(squares, linewidth = 5)
plt.xlabel('Awaeen', fontsize = 14)
plt.ylabel('Bakwas', fontsize = 15)
plt.tick_params('both', direction='out', length=6, width=2, colors='b',
                       grid_color='b', grid_alpha=0.5)
plt.show()
```
